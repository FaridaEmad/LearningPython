```python
stars_container = driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[3]/div[2]/ul")
    
# Find all <li> elements with the class "active"
active_stars = stars_container.find_elements(By.CSS_SELECTOR, "li.active")
    
# Count the number of active <li> elements
star_count = len(active_stars)
print(f"Star Rating: {star_count}")
```

```python
while True:
    try:
        load_more_button = driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/button")
        driver.execute_script("arguments[0].scrollIntoView();", load_more_button)
        load_more_button.click()
        WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[*]"))
            )
        revs = driver.find_elements(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[*]")
        print("Loaded reviews count:", len(revs))
    except NoSuchElementException:
        print("No more reviews to load. The button is no longer available.")
        break
    if(len(revs) > 100):
        break
```

```python
names = []  # Initialize an empty list to store customer names

try:
    # Locate all customer name elements
    cust_names = driver.find_elements(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[*]/div[1]/div/a[1]/h4")
    
    # Loop through each customer name element
    for cust in cust_names:
        try:
            # Extract the text and append to the list
            names.append(cust.text)
            print(f"Customer Name: {cust.text}")
        except Exception as e:
            print(f"Error extracting name for an element: {e}")
            names.append("")  # Append an empty string if there's an error

except Exception as e:
    print(f"Error locating customer name elements: {e}")

print("Final Customer Names List:", names)
```

```python
reviews_list = []  # Initialize an empty list to store reviews

try:
    # Locate all review elements
    reviews = driver.find_elements(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[*]/div[2]/div[1]/p")
    
    # Loop through each review element
    for review in reviews:
        try:
            # Extract the text and append it to the list
            reviews_list.append(review.text)
            print(f"Review: {review.text}")
        except Exception as e:
            print(f"Error extracting review for an element: {e}")
            reviews_list.append("")  # Append an empty string if there's an error

except Exception as e:
    print(f"Error locating review elements: {e}")

print("Final Reviews List:", reviews_list)
```

```python
stars = []  # Initialize the list to store star counts

try:
    # Locate all star containers
    stars_containers = driver.find_elements(By.XPATH, "/html/body/div[3]/div[2]/div[3]/div[5]/div/div[2]/div/div/div/div[2]/div[*]/div[2]/ul")
    
    for container in stars_containers:
        try:
            # Find all active stars in the current container
            active_stars = container.find_elements(By.CSS_SELECTOR, "li.active")
            
            # Count the active stars
            star_count = len(active_stars)
            
            # Append the count to the stars list
            stars.append(star_count)
            print(f"Star count for container: {star_count}")
        except Exception as e:
            print(f"Error processing a star container: {e}")
            stars.append(0)  # Append 0 in case of an error for that container

except Exception as e:
    print(f"Error locating star containers: {e}")

print(stars)
```

```python
# Ensure all lists have the same length
if len(names) == len(stars) == len(reviews_list):
    # Create a dictionary to form the DataFrame
    data = {
        "Name": names,
        "Stars": stars,
        "Review": reviews_list
    }

    # Create the DataFrame
    df = pd.DataFrame(data)

    # Save the DataFrame as a CSV file
    df.to_csv(f"{Rest_name}_reviews.csv", index=False)
    print("DataFrame saved as 'restaurant_reviews.csv'")
else:
    print("Error: Lists have different lengths. Ensure all lists have the same number of elements.")
```
