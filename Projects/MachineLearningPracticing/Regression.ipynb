```python
# Any feature : x 'Independent value' , Output : y 'Dependent value'
# y = @1.x1 + @2.x2 + ....+ @n.xn   '@:Theta'
#      @ between -1,1 'could be negative or positive colleration' 
#      @ : wieght
# 1- Building your equation
# 2- Randamize your weights
#      a- calculating output using the current wieghts
#      b- calculate cost function 'indecates the diffrence between predicted value and real value'
#      c- use GD 'Gradient Descent' to update Thetas '@' depending on the cost function
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('Airline_Delay_Cause.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.drop(['carrier','carrier_name','airport','airport_name'],axis=1,inplace=True)
```

```python
data.info()
```

```python
x = data.drop(['late_aircraft_ct'],axis=1,inplace=False)
y = data['late_aircraft_ct']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=44, shuffle=True)

print('X train shape is: ',x_train.shape)
print('X test shape is: ',x_test.shape)
print('Y train shape is: ',y_train.shape)
print('Y test shape is: ',y_test.shape)
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True,copy_X=True,n_jobs=-1)
```

```python
LinearRegressionModel.fit(x_train,y_train)
```

```python
print('Linear Regression Train Score is : ',LinearRegressionModel.score(x_train,y_train)) #not accurate so do not use it
```

```python
y_pred = LinearRegressionModel.predict(x_test)
list(y_pred)
```

```python
list(y_test)
```

```python
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') #multioutput can be = raw_values
print("Mean Squared Error Value is : ",MSEValue)
```
