```python
data['TweetHour'] = data['created_at'].apply(lambda x : x.split()[3].split(':')[0])
```

```python
data.head()
```

```python
data['TweetHour'].value_counts()
```

```python
def GetDayPartFromHour(n):
    n = int(n)
    if n<=6:
        return 'Early Morning'
    elif n<=10:
        return 'Morning'
    elif n<=13:
        return 'Noon'
    elif n<=17:
        return 'Evening'
    else:
        return 'Night'
```

```python
data['DayPart'] = data['TweetHour'].apply(lambda x: GetDayPartFromHour(x))
```

```python
data.head()
```

```python
def GetDayNameFromDate(Y,M,D):
    wholeDate = str(Y)+'-'+str(M)+'-'+str(D)
    DayName = pd.Timestamp(wholeDate).day_name()
    return DayName

data['NewNameOfDay'] = data.apply(lambda x: GetDayNameFromDate(x['TweetYear'],x['TweetMonth'],x['TweetDay']) ,axis=1)
data.head()
```
