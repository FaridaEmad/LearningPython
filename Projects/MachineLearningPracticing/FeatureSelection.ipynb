```python
import pandas as pd
```

```python
data = pd.read_csv('Airline_Delay_Cause.csv')
```

```python
data
```

```python
def Classes(x):
    if x<=70:
        return 'A'
    elif x < 100:
        return 'B'
    else:
        return 'C'

data['ArrClasses'] = data['arr_flights'].apply(lambda x : Classes(x))
data['ArrClasses'].value_counts()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(['year', 'month', 'carrier', 'carrier_name', 'airport', 'airport_name'],axis=1,inplace=True)
data.head()
```

```python
x = data.drop(['ArrClasses'],axis=1)
y = data['ArrClasses']
```

```python
from sklearn.feature_selection import SelectPercentile ,chi2, f_classif
```

```python
FeatureSelection = SelectPercentile(score_func= f_classif,percentile=50) # chi2 does not work with negative values
```

```python
x_selected = FeatureSelection.fit_transform(x,y)
pd.DataFrame(x_selected)
```

```python
FeatureSelection.get_support()
```

```python
x.columns
```

```python
pd.DataFrame(x_selected,columns=[i for i,j in zip(x.columns,FeatureSelection.get_support()) if j])
```

```python
def FeatureSelectionByPercentile(x,y,percent):
    FeatureSelection = SelectPercentile(score_func= f_classif,percentile=percent)
    x_selected = FeatureSelection.fit_transform(x,y)
    newData=pd.DataFrame(x_selected,columns=[i for i,j in zip(x.columns,FeatureSelection.get_support()) if j])
    newData['y']=y
    return newData

NewData = FeatureSelectionByPercentile(x,y,20)
NewData.head()
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
def FeatureSelectionByNumber(x,y,number):
    FeatureSelection = SelectKBest(score_func= f_classif,k=number)
    x_selected = FeatureSelection.fit_transform(x,y)
    newData=pd.DataFrame(x_selected,columns=[i for i,j in zip(x.columns,FeatureSelection.get_support()) if j])
    newData['y']=y
    return newData

NewData = FeatureSelectionByNumber(x,y,5)
NewData.head()
```
