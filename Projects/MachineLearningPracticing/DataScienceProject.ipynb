```python
data['Formation'].value_counts()
```

```python
data = data[data['Formation'].apply(lambda x: not ('2001' in x or '2002' in x or '2003' in x or '◆' in x))]
data['Formation'].value_counts()
```

```python
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
data
```

```python
unseenData = data[360:]
unseenData
```

```python
data.drop(unseenData.index, axis=0, inplace=True)
data
```

```python
data['Result'].value_counts()
```

```python
draws = data[data['Result'] == 'D']
wins = data[data['Result'] == 'W']
losses = data[data['Result'] == 'L']
draws
```

```python
draws_sampled = draws.sample(n=15)
draws_sampled
```

```python
imbalanced_data = pd.concat([wins, losses, draws_sampled])
imbalanced_data
```

```python
data = imbalanced_data.sample(frac=1, random_state=42).reset_index(drop=True)
data.head()
```
