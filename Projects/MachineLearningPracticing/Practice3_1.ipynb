```python
def StringToDict(f):
    f = f.strip('"[{').strip('}]"')
    f = f.split('}, {')
    FinalList = []
    for item in f:
        NewList = [i for i in [i.strip().split(':') for i in item.split(',')]]
        FinalList.append({i[0].replace("'",""):i[1] for i in NewList if len(i)>1})
    return FinalList

f = credit.iloc[0,0]
g = StringToDict(f)
g
```
